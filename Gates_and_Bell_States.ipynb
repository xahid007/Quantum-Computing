## 2 Qubit Gates

### CX Gate

The Controlled-Not gate, commonly referred to as the CNOT gate, is a fundamental two-qubit gate in quantum computing. It is a key component in quantum circuits, serving as a building block for more complex quantum operations and algorithms.

### Description of the CNOT Gate

The CNOT gate operates on two qubits:
- The **control qubit** (usually the first qubit).
- The **target qubit** (usually the second qubit).

The action of the CNOT gate can be described as follows:
- If the control qubit is in state $|0\rangle$, the target qubit remains unchanged.
- If the control qubit is in state $|1\rangle$, the target qubit is flipped (i.e., if it is $|0\rangle$, it becomes $|1\rangle$; if it is $|1\rangle$, it becomes $|0\rangle$.


```python
# Create the circuit with 2 qbits
circuit = QuantumCircuit(2)

# Apply the CNOT gate
circuit.cx(0,1)  #(control, target)

# Draw the circuit
circuit.draw(output = 'mpl', initial_state = True)
```

## 3 Qubit Gates

### CCX

Here is the description of the CCX gate formatted for Markdown, including inline and display LaTeX:

---

### Description of the CCX (Toffoli) Gate

The CCX gate, also known as the Toffoli gate or controlled-controlled-X gate, is a three-qubit gate in quantum computing. It is an essential gate for constructing more complex quantum circuits and is crucial for certain quantum algorithms, especially those involving reversible computation and quantum error correction.

The CCX gate operates on three qubits:
- Two **control qubits** (typically the first and second qubits).
- One **target qubit** (typically the third qubit).

The action of the CCX gate can be described as follows:
- If both control qubits are in the state $1\rangle$, the target qubit is flipped (i.e., if it is $|0\rangle$, it becomes $1\rangle$; if it is $|1\rangle$, it becomes $|0\rangle$.
- If either or both control qubits are in the state $0\rangle$, the target qubit remains unchanged.


```python
# Create a circuit with 3 qubits
circuit = QuantumCircuit(3)

# Apply ccx gate
circuit.ccx(0,1,2) #(control 1, control 2, target)

# Draw the circuit
circuit.draw(output = 'mpl', initial_state = True)
```

## Simulator

### State Vector

```python
# Build a quantum circuit with one qubit
circuit = QuantumCircuit(1)

# Apply hadamard gate
circuit.h(0)

# Draw the circuit
circuit.draw(output = 'mpl', initial_state = True)
```

```python
#select simulator from Aer package
simulator = Aer.get_backend('aer_simulator')

# Save the statevector
circuit.save_statevector()

circuit = transpile(circuit,simulator)

# Run the simulator
result = simulator.run(circuit).result()
statevector = result.get_statevector(circuit)

print(statevector)
array_to_latex(statevector) #vector representation
```

```python
# Visualize using bloch sphere
plot_bloch_multivector(statevector)
```

### Result after measurement

```python
# Create a circuit with one qubit and one cbit to keep the measurement
circuit = QuantumCircuit(1,1)

# Apply hadamard game
circuit.h(0)

# Measurement
circuit.measure(0,0)

# Draw the circuit
circuit.draw(output = 'mpl', initial_state = True)
```

```python
# Select the simulator package from Aer
simulator = Aer.get_backend('aer_simulator')

# Adapt the circuit
circuit = transpile(circuit, simulator)

# Run the circuit
job = simulator.run(circuit, shots = 1000)

# Monitor the job
job_monitor(job)

# Get the result
result = job.result()

# count the result
count = result.get_counts()

# Plot in histogram
plot_histogram(count)
```

## Bell States
- $$|\Phi^+\rangle = \frac{1}{\sqrt{2}} (|00\rangle + |11\rangle)$$
- $$|\Phi^-\rangle = \frac{1}{\sqrt{2}} (|00\rangle - |11\rangle)$$
- $$|\Psi^+\rangle = \frac{1}{\sqrt{2}} (|01\rangle + |10\rangle)$$
- $$|\Psi^-\rangle = \frac{1}{\sqrt{2}} (|01\rangle - |10\rangle)$$

### Bell state 1

```python
# Create a circuit with two qbits
circuit = QuantumCircuit(2)

# Apply hadamard gate on the first qbit
circuit.h(0)

# Apply CNOT gate (first qbit is the control and the second one is the target)
circuit.cx(0,1)

# Draw the circuit
circuit.draw(output = 'mpl', initial_state = True)
```

```python
# select the simulator package from Aer
simulator = Aer.get_backend('aer_simulator')

# Save the state vector
circuit.save_statevector()

# Adapt to the simulator
circuit = transpile(circuit, simulator)

# Run the circuit and get result
result = simulator.run(circuit).result()

# Get the state vector
statevector = result.get_statevector(circuit)

count = result.get_counts()
print(statevector)
array_to_latex(statevector)
```

```python
plot_bloch_multivector(statevector)
```

```python
plot_histogram(count)
```
